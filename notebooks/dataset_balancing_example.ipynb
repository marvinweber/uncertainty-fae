```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()

df_train = pd.read_csv('/data_fae_uq/rsna_boneage/train_annotations.csv')
df_val = pd.read_csv('/data_fae_uq/rsna_boneage/val_annotations.csv')
df_test = pd.read_csv('/data_fae_uq/rsna_boneage/test_annotations.csv')

for df in [df_train, df_val, df_test]:
    df['sex'] = df.apply(lambda row: 'M' if row[['male']].bool() else 'F', axis=1)
    df['boneage_years'] = df['boneage'] / 12

df_all = pd.concat([df_train, df_val, df_test])
df_all[['id', 'boneage', 'boneage_years', 'sex']]
```

## Unbalanced Raw (Full) Dataset

```python
genders = df_all.sex.unique()
fig, ax = plt.subplots(figsize=(8,5))
ax.hist(
    [df_all.loc[df_all.sex == x, 'boneage_years'] for x in genders],
    label=genders,
    bins=10,
)
ax.set_xlabel('Age (Years)')
ax.legend()
fig.show()
```

## Unbalanced Train Dataset

```python
genders = df_train.sex.unique()
fig, ax = plt.subplots(figsize=(8,5))
ax.hist(
    [df_train.loc[df_train.sex == x, 'boneage_years'] for x in genders],
    label=genders,
    bins=10,
)
ax.set_xlabel('Age (Years)')
ax.legend()
fig.show()
```

## Rebalanced Train Dataset

```python
df_train['boneage_bin'] = pd.cut(df_train['boneage'], 10)
df_train_balanced = df_train.groupby(
    ['boneage_bin', 'sex']
).apply(
    lambda x: x.sample(1100, replace = True)
).reset_index(drop = True)

print('New Data Size:', df_train_balanced.shape[0], 'Old Size:', df_train.shape[0])

genders = df_train_balanced.sex.unique()
fig, ax = plt.subplots(figsize=(8,5))
ax.hist(
    [df_train_balanced.loc[df_train_balanced.sex == x, 'boneage_years'] for x in genders],
    label=genders,
    bins=10,
)
ax.set_xlabel('Age (Years)')
ax.legend()
fig.show()
```
