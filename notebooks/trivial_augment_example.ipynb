```python
from torchvision import transforms
from PIL import Image

img_path = "/data_fae_uq/rsna_boneage/train_val/4184.png"
resize_transform = transforms.Resize(200)
transform = transforms.Compose(
    [
        transforms.TrivialAugmentWide(),
        resize_transform,
    ]
)

image = Image.open(img_path).convert("RGB")
resize_transform(image).show()

for _ in range(10):
    image_transformed = transform(image)
    image_transformed.show(title="Random")
```
